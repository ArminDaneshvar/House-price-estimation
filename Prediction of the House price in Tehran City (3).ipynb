# Prediction of the House price in Tehran City

```python
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
from sklearn import linear_model 
```

```python
df=pd.read_csv('Reg.csv')
```

```python
pd.DataFrame(df)
```

```python
df['Area'] = df['Area'].str.replace(',', '').astype(float)
df['Area'] = pd.to_numeric(df.Area)
```

```python
df["Parking"] = df["Parking"].astype(int)
df["Warehouse"] = df["Warehouse"].astype(int)
df["Elevator"] = df["Elevator"].astype(int)
new_df = df.dropna()
```

```python
new_dff=new_df['Area'].idxmax()
new_df_cl=new_df['Area']
new_df_cl
```

```python
max=[]
for i in range(3456):
    if  new_df['Area'].max()>7000: 
        new_dfff=new_df['Area'].idxmax()
        max.append(new_dfff)
        new_df=new_df.drop(index=[max[i]])

    
new_df['Area'].max()
new_df

       
```

```python
cdf=new_df[['Area','Room', 'Parking', 'Elevator', 'Warehouse','Price(USD)']]
msk=np.random.rand(len(new_df))<0.85
train=cdf[msk]
test=cdf[~msk]
regr=linear_model.LinearRegression()
x=np.asanyarray(train[['Area','Room', 'Parking', 'Warehouse','Elevator']])
y=np.asanyarray(train[['Price(USD)']])
regr.fit(x,y)
print('coefficents: ', regr.coef_)
print('intercept: ', regr.intercept_)
```

```python
yhat=regr.predict(test[['Area','Room', 'Parking', 'Warehouse','Elevator']])
x =np.asanyarray(test[['Area','Room', 'Parking','Warehouse', 'Elevator']])
y =np.asanyarray(test[['Price(USD)']])
print('v squre:' ,regr.score(x,y))
```
